### Method 1: Downloading the Dataset to your local machine

```python
#Running dotenv package for first time? Run on console: "pip install python-dotenv"

from dotenv import load_dotenv
import os

load_dotenv(r"E:\Example_Folder\Example_Project\.env")

import kaggle
kaggle.api.authenticate()

# API Call
kaggle.api.dataset_download_files(
    #username and dataset name from the link: https://www.kaggle.com/datasets/tavoglc/dengue-sequences-metadata
    "tavoglc/dengue-sequences-metadata",
    #file path where we wish to place the data in our local machine (creates a folder with the data)
    path=r"E:\Example_Folder\Example_Project\dengue_data",
    #unzip is unzipping the data
    unzip=True
)

print("Download complete!")
```

```python
#Load the dataset after downloading
import pandas as pd

# List downloaded files
import os
files = os.listdir(r"E:\Example_Folder\Example_Project\dengue_data")
print(files)

# Load the CSV (look in the file to see what the data file name is called)
df = pd.read_csv(r"E:\Example_Folder\Example_Project\metdata.csv")
print(df.head())
```

### Method 2: Using Memory instead of downloading dataset

##### Using only memory without downloading. (keep in mind this has to run every time you re-run python)

```python
#Loading the dataset using only memory without downloading. (keep in mind this has to run every time you re-run python)

from dotenv import load_dotenv
import os
import kaggle
import zipfile
import io
import pandas as pd

load_dotenv(r"E:\Example_Folder\Example_Project\.env")
kaggle.api.authenticate()

# Download to memory instead of disk
zip_data = kaggle.api.dataset_download_files(
    "tavoglc/dengue-sequences-metadata",
    #creates a folder called temp with it's data on it
    path=r"E:\Example_Folder\Example_Project\temp",
    unzip=False
)

# Read the zip file without extracting the files
with zipfile.ZipFile(r"E:\Example_Folder\Example_Project\temp\dengue-sequences-metadata.zip", "r") as z:
    print(z.namelist())  # See what files are inside
    with z.open(z.namelist()[0]) as f:  # Open first file
        df = pd.read_csv(f)

print(df.head())
```
